# Splitting The data

```python
from sklearn.model_selection import train_test_split
#Defining the features and target
x = df.drop(['animal name','type'],axis = 1) # Dropping the animal name col as we do not want to know the animal name for the model when the name is known predicting is easier.
y =df['type']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
print('Shape of x_train:',x_train.shape)
print('Shape of x_test:',x_test.shape)
print('Shape of y_train:',y_train.shape)
print('Shape of y_test:',y_test.shape)
```

# Scaling data using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

## Finding the best k(nearest neighbor)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
k_range = range(1,15)
accuracy_scores = []
for k in k_range:
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(x_train,y_train)
  y_pred = knn.predict(x_test)
  acc = accuracy_score(y_test,y_pred)
  accuracy_scores.append(acc)
  print('Accuracy for k = {} is {:.2f}'.format(k,acc))
```

# Elbow curve

```python
# Plotting k_range and accuracy scores to check the elbow and decide best k value
plt.figure(figsize = (10,6))
plt.plot(k_range,accuracy_scores,marker = 'o')
plt.title('K-value vs Accuracy')
plt.xlabel('k-value')
plt.ylabel('accuracy score')
plt.show()
```

# Model Building

```python
best_k = 3
# We will choose Euclidean as our distance metric as it's a strong, standard choice.
best_distance_metric = 'euclidean'

knn_model = KNeighborsClassifier(n_neighbors=best_k,metric = best_distance_metric )
knn_model.fit(x_train,y_train)
y_pred = knn_model.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print('Accuracy for k = {} is {:.2f}'.format(best_k,acc))
print(f"\nKNN classifier implemented with K={best_k} and {best_distance_metric} as distance metric.")
```

# Model Evaluation

```python
from sklearn.metrics import classification_report


y_pred = knn_model.predict(x_test)
print("--- Model Evaluation Report ---")
print(classification_report(y_test, y_pred))
```

# Visualizing the decision boundary

```python
from sklearn.decomposition import PCA
#Implementing pca to to reduce the features(The dataset has nearly 18 features)
pca = PCA(n_components = 2)
x_train_pca = pca.fit_transform(x_train)

# Applying KNN on the pca transformed data(x_train_pca)
knn_pca = KNeighborsClassifier(n_neighbors = best_k)
knn_pca.fit(x_train_pca,y_train)

# Creating a meshgrid to plot the decision boundary
x_min,x_max = x_train_pca[:,0].min() -1,x_train_pca[:,0].max() +1
y_min,y_max = x_train_pca[:,1].min() -1,x_train_pca[:,1].max() +1
# -1 and +1 is the min and max value given to mweshgrid so the points do  not touch the edges or the points go out of the plot range
xx,yy = np.meshgrid(np.arange(x_min,x_max,0.1),np.arange(y_min,y_max,0.1)) # Here a meshgrid is created with min and max value of x and y axis


z = knn_pca.predict(np.c_[xx.ravel(),yy.ravel()]) # class of the  point(x,y) is  predicted .ravel() converts 2D array to 1D formed due to meshgrid
Z = z.reshape(xx.shape)

# plotting the decision boundary
plt.figure(figsize = (12,8))
plt.contourf(xx,yy,Z,alpha = 0.4,cmap = 'viridis')
scatter = plt.scatter(x_train_pca[:,0],x_train_pca[:,1],c=y_train,s = 30,edgecolor = 'k',cmap = 'viridis')
unique_classes = sorted(y_train.unique())
colors = plt.cm.get_cmap('viridis',len(unique_classes))
# Create a legend
class_names = [str(i) for i in sorted(y_train.unique())]
plt.legend(handles=scatter.legend_elements()[0], labels=class_names, title="Animal Types")
plt.title(f'KNN Decision Boundaries with PCA (K={best_k})')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()


```

# Key Hyperparameters used in KNN

Hyperparameters of KNN

n_neighbors → Number of neighbors (k) to consider.

weights → How neighbors influence (uniform / distance).

metric → Distance function (Euclidean, Manhattan, etc.).

p → Power for Minkowski (p=1 → Manhattan, p=2 → Euclidean).

algorithm → Method for neighbor search (auto, ball_tree, kd_tree, brute).

leaf_size → Efficiency parameter for tree-based search.

n_jobs → Number of parallel CPU cores to use.

# Distance metrics used in KNN

KNN can use various distance metrics such as Euclidean, Manhattan, Minkowski, Chebyshev, Cosine, and Hamming distance, depending on the type of data and problem.
